```python
import json
import os

import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_dir = './assets/results'
files = [f for f in os.listdir(data_dir) if f.endswith('.json')]
```

```python
all_data = {}

for file in files:
    method = file.split('.')[0]
    with open(os.path.join(data_dir, file), 'r') as f:
        data = json.load(f)
        all_data[method] = data

nodes = set()
for method_data in all_data.values():
    nodes.update(method_data['cpu'].keys())
nodes = sorted(list(nodes))
```

```python
cpu_dfs = []
mem_dfs = []

for method, data in all_data.items():
    cpu = pd.DataFrame(data['cpu'], index=[method]).T
    cpu.columns = [method]
    cpu_dfs.append(cpu)
    mem = pd.DataFrame(data['mem'], index=[method]).T
    mem.columns = [method]
    mem_dfs.append(mem)

cpu_df = pd.concat(cpu_dfs, axis=1).reindex(nodes)
mem_df = pd.concat(mem_dfs, axis=1).reindex(nodes)
```

```python
def get_uniform_distribution(df):
    total = df.sum().mean()
    uniform_value = total / len(df)
    return pd.Series([uniform_value] * len(df), index=df.index, name='uniform')


cpu_df_with_uniform = cpu_df.copy()
cpu_df_with_uniform['uniform'] = get_uniform_distribution(cpu_df)
mem_df_with_uniform = mem_df.copy()
mem_df_with_uniform['uniform'] = get_uniform_distribution(mem_df)
```

```python
# 1. Столбчатый график с линией равномерного распределения
plt.figure(figsize=(16, 6))
cpu_df.plot(kind='bar', figsize=(16, 6))
plt.plot(cpu_df.index, cpu_df_with_uniform['uniform'], linestyle='--', marker='o', label='Равномерное распределение',
         linewidth=2)
plt.title('Распределение CPU по методам (с равномерным распределением)')
plt.xlabel('Нода')
plt.ylabel('CPU')
plt.xticks(rotation=45)
plt.legend(title='Метод')
plt.tight_layout()
plt.show()

plt.figure(figsize=(16, 6))
mem_df.plot(kind='bar', figsize=(16, 6))
plt.plot(mem_df.index, mem_df_with_uniform['uniform'], linestyle='--', marker='o', label='Равномерное распределение',
         linewidth=2)
plt.title('Распределение MEM по методам (с равномерным распределением)')
plt.xlabel('Нода')
plt.ylabel('MEM')
plt.xticks(rotation=45)
plt.legend(title='Метод')
plt.tight_layout()
plt.show()
```

```python
# 2. Boxplot для каждого метода + равномерное распределение
plt.figure(figsize=(10, 6))
cpu_df_with_uniform.boxplot()
plt.title('Boxplot CPU по методам и равномерному распределению')
plt.ylabel('CPU')
plt.show()

plt.figure(figsize=(10, 6))
mem_df_with_uniform.boxplot()
plt.title('Boxplot MEM по методам и равномерному распределению')
plt.ylabel('MEM')
plt.show()
```

```python
data_dir = './assets/results_with_latency'
files = [f for f in os.listdir(data_dir) if f.endswith('.json')]
metrics_data = {}

for file in files:
    method = file.split('.')[0]
    with open(os.path.join(data_dir, file), 'r') as f:
        data = json.load(f)
        metrics_data[method] = data

nodes = set()
for method_data in metrics_data.values():
    nodes.update(method_data['cpu'].keys())
nodes = sorted(list(nodes))
```

```python
total_df = pd.DataFrame(
    {
        alg: data["total_makespan_s"]
        for alg, data in metrics_data.items()
    },
    index=["total_makespan_s"]
).T  # algorithms as rows

# show interactive table

# bar chart
plt.figure()
plt.bar(total_df.index, total_df["total_makespan_s"])
plt.ylabel("Total makespan (s)")
plt.title("Total makespan comparison")
plt.tight_layout()
plt.show()

# ---------------------------------------------------------------------------
# 2) Average latency per port (one chart per algorithm)
for alg, data in metrics_data.items():
    ports  = list(data["avg_latency_ms_by_port"].keys())
    values = list(data["avg_latency_ms_by_port"].values())

    plt.figure()
    plt.bar(ports, values)
    plt.xticks(rotation=45)
    plt.ylabel("Avg latency (ms)")
    plt.title(f"Avg latency per port – {alg}")
    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------------------------
# 3) Per‑wave makespan (all algorithms on one chart)
plt.figure()
for alg, data in metrics_data.items():
    waves  = list(range(1, len(data["per_wave_makespan_s"]) + 1))
    values = data["per_wave_makespan_s"]
    plt.plot(waves, values, label=alg)

plt.xlabel("Wave #")
plt.ylabel("Makespan per wave (s)")
plt.title("Per‑wave makespan")
plt.legend()
plt.tight_layout()
plt.show()

# ---------------------------------------------------------------------------
# 4) CPU average % per port (one chart per algorithm)
for alg, data in metrics_data.items():
    ports  = list(data["cpu_avg_percentage"].keys())
    values = [v * 100 for v in data["cpu_avg_percentage"].values()]  # convert to %

    plt.figure()
    plt.bar(ports, values)
    plt.xticks(rotation=45)
    plt.ylabel("CPU avg (%)")
    plt.title(f"CPU avg % per port – {alg}")
    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------------------------
# 5) Memory average % per port (one chart per algorithm)
for alg, data in metrics_data.items():
    ports  = list(data["mem_avg_percentage"].keys())
    values = [v * 100 for v in data["mem_avg_percentage"].values()]  # convert to %

    plt.figure()
    plt.bar(ports, values)
    plt.xticks(rotation=45)
    plt.ylabel("Memory avg (%)")
    plt.title(f"Memory avg % per port – {alg}")
    plt.tight_layout()
    plt.show()
```
